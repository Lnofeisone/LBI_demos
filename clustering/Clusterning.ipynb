```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
sns.set()
import numpy as np
%matplotlib inline
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.6, 
                  random_state = 0)
x = X[:, 0]
y = X[:, 1]
plt.scatter(x, y, s = 30, marker = "o", color = ("#86BC25"))
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
#y_kmeans.size
```

```python
plt.scatter(x, y, c = y_kmeans, s = 50, cmap = 'Spectral')
```

```python
plt.scatter(x, y, c=y_kmeans, s = 50, cmap = 'Spectral')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = "red", s = 100, alpha = 1)
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)

labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='prism');
```

```python
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=2, affinity='rbf',
                           assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='prism');
```
